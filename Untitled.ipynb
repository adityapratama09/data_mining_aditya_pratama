```python
# -*- coding: utf-8 -*-
"""
# ANALISIS ASSOCIATION RULE MINING PADA DATASET SUPERMARKET
## Menggunakan Framework CRISP-DM

Nama : Aditya Pratama
NIM : 1224160041
Kelas : SI 24 P SIM 2
"""

# =============================================================================
# A. Business Understanding
# =============================================================================
"""
## Latar Belakang Masalah

Perusahaan retail menghasilkan data transaksi penjualan dalam jumlah besar setiap harinya. 
Data tersebut menyimpan berbagai informasi penting mengenai perilaku pelanggan dalam 
melakukan pembelian produk.

Namun, sebagian besar data transaksi hanya digunakan sebagai arsip tanpa dilakukan 
analisis lebih lanjut. Padahal data tersebut dapat dimanfaatkan untuk menemukan pola 
pembelian pelanggan sehingga dapat membantu perusahaan dalam meningkatkan strategi 
pemasaran dan penjualan.

Association Rule Mining merupakan salah satu teknik Data Mining yang digunakan untuk 
menemukan hubungan antar item yang sering muncul secara bersamaan dalam suatu transaksi.

## Tujuan Project

1. Menemukan pola hubungan antar produk yang sering dibeli bersamaan.
2. Menghasilkan aturan asosiasi menggunakan algoritma Apriori.
3. Memberikan rekomendasi strategi pemasaran berdasarkan pola pembelian pelanggan.

## Manfaat Analisis

1. Membantu strategi cross-selling.
2. Membantu product bundling.
3. Mengoptimalkan tata letak produk.
4. Meningkatkan penjualan perusahaan.
"""

# =============================================================================
# B. Import Library
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

import warnings
warnings.filterwarnings('ignore')

# Set style untuk visualisasi
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ Library berhasil diimport!")

# =============================================================================
# C. Data Understanding
# =============================================================================

print("\n" + "="*80)
print("DATA UNDERSTANDING")
print("="*80)

# Membaca dataset
df = pd.read_csv('supermarket.csv')

print("\n1. 5 Data Teratas:")
print(df.head())

print("\n2. Informasi Dataset:")
print(df.info())

print("\n3. Statistik Deskriptif:")
print(df.describe(include='all'))

print("\n4. Cek Missing Values:")
print(df.isnull().sum())

print("\n5. Distribusi Kategori:")
print(df['Category'].value_counts())

print("\n6. Distribusi Sub-Category:")
print(df['Sub-Category'].value_counts().head(10))

print("\n7. Distribusi Segment:")
print(df['Segment'].value_counts())

# Visualisasi data
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df['Category'].value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title('Distribusi Kategori Produk')
axes[0].set_xlabel('Kategori')
axes[0].set_ylabel('Jumlah Transaksi')
axes[0].tick_params(axis='x', rotation=45)

df['Segment'].value_counts().plot(kind='bar', ax=axes[1])
axes[1].set_title('Distribusi Segment Pelanggan')
axes[1].set_xlabel('Segment')
axes[1].set_ylabel('Jumlah Transaksi')
axes[1].tick_params(axis='x', rotation=45)

df['Region'].value_counts().plot(kind='bar', ax=axes[2])
axes[2].set_title('Distribusi Wilayah')
axes[2].set_xlabel('Region')
axes[2].set_ylabel('Jumlah Transaksi')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Top Sub-Category
top_sub = df['Sub-Category'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_sub.plot(kind='bar')
plt.title('Top 10 Sub-Category Terlaris')
plt.xlabel('Sub Category')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.show()

# =============================================================================
# D. Data Preparation
# =============================================================================

print("\n" + "="*80)
print("DATA PREPARATION")
print("="*80)

# 1. Data Cleaning - Hapus missing values
df = df.dropna()
print(f"Jumlah Data Setelah Cleaning : {df.shape}")

# 2. Buat transaksi untuk market basket analysis
transactions = df.groupby('Order ID')['Product Name'].apply(list).tolist()
print(f"\nJumlah transaksi: {len(transactions)}")

# 3. Encoding Data dengan TransactionEncoder
print("\n🧩 Encoding Data dengan TransactionEncoder")
print("-" * 50)

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print(f"✅ Encoding selesai!")
print(f"📐 Shape data encoded: {df_encoded.shape}")
print("\n🔍 Sample Data Encoded (5 transaksi pertama):")
print(df_encoded.head())

# =============================================================================
# E. Modeling - Algoritma Apriori
# =============================================================================

print("\n" + "="*80)
print("MODELING - ALGORITMA APRIORI")
print("="*80)

min_support = 0.05
print(f"📌 Parameter:")
print(f" - Min. Support: {min_support * 100}%")
print(f" - Total Transaksi: {len(df_encoded)}")
print(f" - Min. Frekuensi: {int(min_support * len(df_encoded))} transaksi")

print("\n⏳ Menjalankan algoritma Apriori...")

frequent_itemsets_apriori = apriori(
    df_encoded,
    min_support=min_support,
    use_colnames=True
)

print(f"✅ Selesai! Ditemukan {len(frequent_itemsets_apriori)} frequent itemsets")

frequent_itemsets_apriori = frequent_itemsets_apriori.sort_values(
    'support', ascending=False
)

print("\n📋 Top 20 Frequent Itemsets:")
print(frequent_itemsets_apriori.head(20))

# Generate Association Rules
print("\n🔗 Generate Association Rules (Apriori)")
print("-" * 50)

min_confidence = 0.50
print(f"📌 Parameter:")
print(f" - Min. Confidence: {min_confidence * 100}%")

print("\n⏳ Generating association rules...")

rules_apriori = association_rules(
    frequent_itemsets_apriori,
    metric='confidence',
    min_threshold=min_confidence
)

print(f"✅ Selesai! Ditemukan {len(rules_apriori)} association rules")

rules_apriori['antecedent_len'] = rules_apriori['antecedents'].apply(lambda x: len(x))
rules_apriori['consequent_len'] = rules_apriori['consequents'].apply(lambda x: len(x))

rules_apriori_filtered = rules_apriori[rules_apriori['lift'] > 1].copy()
print(f"\n📊 Rules dengan Lift > 1: {len(rules_apriori_filtered)}")

rules_apriori_filtered = rules_apriori_filtered.sort_values(
    ['lift', 'confidence', 'support'], ascending=False
)

print("\n📋 Top 20 Association Rules (Apriori):")
print(rules_apriori_filtered.head(20))

# FP-Growth
print("\n⚡ Modeling dengan Algoritma FP-Growth")
print("-" * 50)

frequent_itemsets_fpgrowth = fpgrowth(
    df_encoded,
    min_support=min_support,
    use_colnames=True
)

print(f"✅ Selesai! Ditemukan {len(frequent_itemsets_fpgrowth)} frequent itemsets")

frequent_itemsets_fpgrowth = frequent_itemsets_fpgrowth.sort_values(
    'support', ascending=False
)

print("\n📋 Top 20 Frequent Itemsets:")
print(frequent_itemsets_fpgrowth.head(20))

# Generate Rules FP-Growth
print("\n🔗 Generate Association Rules (FP-Growth)")
print("-" * 50)

rules_fpgrowth = association_rules(
    frequent_itemsets_fpgrowth,
    metric='confidence',
    min_threshold=min_confidence
)

print(f"✅ Selesai! Ditemukan {len(rules_fpgrowth)} association rules")

rules_fpgrowth['antecedent_len'] = rules_fpgrowth['antecedents'].apply(lambda x: len(x))
rules_fpgrowth['consequent_len'] = rules_fpgrowth['consequents'].apply(lambda x: len(x))

rules_fpgrowth_filtered = rules_fpgrowth[rules_fpgrowth['lift'] > 1].copy()
print(f"\n📊 Rules dengan Lift > 1: {len(rules_fpgrowth_filtered)}")

rules_fpgrowth_filtered = rules_fpgrowth_filtered.sort_values(
    ['lift', 'confidence', 'support'], ascending=False
)

print("\n📋 Top 20 Association Rules (FP-Growth):")
print(rules_fpgrowth_filtered.head(20))

# =============================================================================
# F. Evaluation
# =============================================================================

print("\n" + "="*80)
print("EVALUASI")
print("="*80)

# Perbandingan Apriori vs FP-Growth
print("\n📊 Perbandingan Hasil: Apriori vs FP-Growth")
print("="*50)

comparison_data = {
    'Metrik': [
        'Frequent Itemsets',
        'Association Rules (Lift>1)',
        'Avg. Confidence',
        'Avg. Lift',
        'Max. Lift'
    ],
    'Apriori': [
        len(frequent_itemsets_apriori),
        len(rules_apriori_filtered),
        f"{rules_apriori_filtered['confidence'].mean():.2%}",
        f"{rules_apriori_filtered['lift'].mean():.2f}",
        f"{rules_apriori_filtered['lift'].max():.2f}"
    ],
    'FP-Growth': [
        len(frequent_itemsets_fpgrowth),
        len(rules_fpgrowth_filtered),
        f"{rules_fpgrowth_filtered['confidence'].mean():.2%}",
        f"{rules_fpgrowth_filtered['lift'].mean():.2f}",
        f"{rules_fpgrowth_filtered['lift'].max():.2f}"
    ]
}

df_comparison = pd.DataFrame(comparison_data)
print(df_comparison)

# Top Association Rules
print("\n🏆 TOP 15 ASSOCIATION RULES")
print("="*50)

top_rules = rules_fpgrowth_filtered.head(15).copy()

for idx, row in top_rules.iterrows():
    antecedents = ', '.join(list(row['antecedents']))
    consequents = ', '.join(list(row['consequents']))
    
    print(f"\n📌 Rule #{len(top_rules) - list(top_rules.index).index(idx)}")
    print(f"   Jika membeli: [{antecedents}]")
    print(f"   Maka akan membeli: [{consequents}]")
    print(f"   Support: {row['support']:.2%} ({row['support']*len(df_encoded):.0f} transaksi)")
    print(f"   Confidence: {row['confidence']:.2%}")
    print(f"   Lift Ratio: {row['lift']:.2f}")
    
    if row['lift'] >= 1.5:
        strength = "🔥 SANGAT KUAT"
    elif row['lift'] >= 1.2:
        strength = "💪 KUAT"
    else:
        strength = "📊 SEDANG"
    
    print(f"   Kekuatan: {strength}")

# Visualisasi Distribusi Metrik
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(rules_fpgrowth_filtered['support'], bins=20, color='steelblue', edgecolor='navy', alpha=0.7)
axes[0].set_title('Distribusi Support', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Support', fontsize=10)
axes[0].set_ylabel('Frekuensi', fontsize=10)
axes[0].grid(axis='y', alpha=0.3)

axes[1].hist(rules_fpgrowth_filtered['confidence'], bins=20, color='forestgreen', edgecolor='darkgreen', alpha=0.7)
axes[1].set_title('Distribusi Confidence', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Confidence', fontsize=10)
axes[1].set_ylabel('Frekuensi', fontsize=10)
axes[1].grid(axis='y', alpha=0.3)

axes[2].hist(rules_fpgrowth_filtered['lift'], bins=20, color='coral', edgecolor='darkred', alpha=0.7)
axes[2].set_title('Distribusi Lift Ratio', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Lift Ratio', fontsize=10)
axes[2].set_ylabel('Frekuensi', fontsize=10)
axes[2].grid(axis='y', alpha=0.3)

plt.suptitle('Distribusi Metrik Evaluasi Association Rules', fontsize=14, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()

# Scatter Plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    x=rules_fpgrowth_filtered['support'] * 100,
    y=rules_fpgrowth_filtered['confidence'] * 100,
    c=rules_fpgrowth_filtered['lift'],
    s=100,
    alpha=0.6,
    cmap='viridis',
    edgecolors='black',
    linewidths=0.5
)
plt.colorbar(scatter, label='Lift Ratio')
plt.xlabel('Support (%)', fontsize=12, fontweight='bold')
plt.ylabel('Confidence (%)', fontsize=12, fontweight='bold')
plt.title('Scatter Plot: Support vs Confidence (Color = Lift)', fontsize=14, fontweight='bold', pad=20)
plt.grid(alpha=0.3, linestyle='--')
plt.axvline(x=5, color='red', linestyle='--', alpha=0.5, label='Min. Support (5%)')
plt.axhline(y=50, color='blue', linestyle='--', alpha=0.5, label='Min. Confidence (50%)')
plt.legend()
plt.tight_layout()
plt.show()

# Heatmap Co-occurrence
produk_columns = df_encoded.columns[:11]
co_occurrence = np.zeros((len(produk_columns), len(produk_columns)))

for i, prod1 in enumerate(produk_columns):
    for j, prod2 in enumerate(produk_columns):
        if i <= j:
            count = ((df_encoded[prod1] == 1) & (df_encoded[prod2] == 1)).sum()
            co_occurrence[i][j] = count
            co_occurrence[j][i] = count

df_co_occurrence = pd.DataFrame(
    co_occurrence,
    index=produk_columns,
    columns=produk_columns
)

plt.figure(figsize=(12, 10))
sns.heatmap(
    df_co_occurrence,
    annot=True,
    fmt='.0f',
    cmap='YlOrRd',
    cbar_kws={'label': 'Jumlah Transaksi'}
)
plt.title('Heatmap: Co-occurrence Produk', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Produk', fontsize=12, fontweight='bold')
plt.ylabel('Produk', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# =============================================================================
# G. Deployment - Rekomendasi Bisnis
# =============================================================================

print("\n" + "="*80)
print("DEPLOYMENT - REKOMENDASI STRATEGI BISNIS")
print("="*80)

# Filter rules dengan kriteria kuat
strong_rules = rules_fpgrowth_filtered[
    (rules_fpgrowth_filtered['lift'] >= 1.3) & 
    (rules_fpgrowth_filtered['confidence'] >= 0.60) & 
    (rules_fpgrowth_filtered['support'] >= 0.05)
].copy()

print(f"\n📄 Ditemukan {len(strong_rules)} rules dengan kriteria ketat.")
print("\n" + "="*50)

rekomendasi_count = 1

for idx, row in strong_rules.iterrows():
    antecedents = list(row['antecedents'])
    consequents = list(row['consequents'])
    
    print(f"\n⭐ REKOMENDASI #{rekomendasi_count}")
    print("-" * 50)
    print(f"📦 Produk Utama: {', '.join(antecedents)}")
    print(f"🤝 Produk Rekomendasi: {', '.join(consequents)}")
    
    print(f"\n📊 Statistik:")
    print(f" - Support: {row['support']:.2%} ({int(row['support']*len(df_encoded))} transaksi)")
    print(f" - Confidence: {row['confidence']:.2%}")
    print(f" - Lift Ratio: {row['lift']:.2f}")
    
    print(f"\n💡 Strategi Implementasi:")
    print(f" 1. BUNDLING: Buat paket promo '{', '.join(antecedents)} + {', '.join(consequents)}' dengan diskon 10-15%")
    print(f" 2. LAYOUT: Letakkan {', '.join(consequents)} dekat dengan {', '.join(antecedents)}")
    print(f" 3. PROMOSI: Tampilkan banner 'Pelanggan yang membeli {', '.join(antecedents)} juga membeli {', '.join(consequents)}'")
    print(f" 4. CROSS-SELL: Kasir dapat menawarkan {', '.join(consequents)} saat pelanggan membeli {', '.join(antecedents)}")
    
    rekomendasi_count += 1

print("\n" + "="*50)
print(f"✅ Total {rekomendasi_count-1} rekomendasi bisnis berhasil dibuat!")
print("="*50)

# =============================================================================
# Ringkasan Eksekutif
# =============================================================================

print("\n" + "="*80)
print("📑 RINGKASAN EKSEKUTIF - ANALISIS ASOSIASI SUPERMARKET")
print("="*80)

print(f"\n🧾 Total Transaksi: {len(transactions):,}")
print(f"🛒 Produk Dianalisis: {len(produk_columns)} item")

print(f"\n⚙️ Model Performance:")
print(f" - Frequent Itemsets: {len(frequent_itemsets_fpgrowth)}")
print(f" - Association Rules (Lift>1): {len(rules_fpgrowth_filtered)}")
print(f" - Avg. Confidence: {rules_fpgrowth_filtered['confidence'].mean():.2%}")
print(f" - Avg. Lift Ratio: {rules_fpgrowth_filtered['lift'].mean():.2f}")

print(f"\n🎯 Rekomendasi Utama:")
print(f" - {len(strong_rules)} strategi bundling produk prioritas")
print(f" - Fokus pada pasangan produk dengan Lift > 1.3")
print(f" - Implementasikan cross-selling di kasir")
print(f" - Optimalkan tata letak toko berdasarkan pola pembelian")

print(f"\n📈 Potensi Dampak Bisnis:")
print(f" - Peningkatan penjualan melalui bundling: Estimasi 10-15%")
print(f" - Peningkatan cross-selling: Estimasi 8-12%")
print(f" - Peningkatan kepuasan pelanggan: Estimasi 5-10%")

print("\n" + "="*80)
print("✅ Analisis selesai! Silakan gunakan rekomendasi untuk strategi bisnis.")
print("="*80)
```

```python
# ============================================
# CEK KESIAPAN DATASET UNTUK ASSOCIATION RULE MINING
# ============================================

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('supermarket.csv')

print("="*60)
print("CEK KESIAPAN DATASET UNTUK ASSOCIATION RULE MINING")
print("="*60)

# ============================================
# 1. CEK STRUKTUR DATA
# ============================================
print("\n【1. CEK STRUKTUR DATA】")
print("-"*40)

print(f"Jumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")
print(f"Kolom yang tersedia: {df.columns.tolist()}")

# Cek apakah ada transaksi (Order ID) dan item (Sub-Category)
wajib_kolom = ['Order ID', 'Sub-Category']
ada_semua = all(kolom in df.columns for kolom in wajib_kolom)
print(f"Kolom 'Order ID' dan 'Sub-Category' tersedia: {'✅ YA' if ada_semua else '❌ TIDAK'}")

if not ada_semua:
    print("⚠️ Data tidak memiliki kolom transaksi/item yang diperlukan!")

# ============================================
# 2. CEK JENIS DATA
# ============================================
print("\n【2. CEK JENIS DATA】")
print("-"*40)
print(df[['Order ID', 'Sub-Category']].dtypes)

# ============================================
# 3. CEK DATA HILANG
# ============================================
print("\n【3. CEK DATA HILANG】")
print("-"*40)
missing_order = df['Order ID'].isna().sum()
missing_sub = df['Sub-Category'].isna().sum()

print(f"Missing pada Order ID: {missing_order} baris")
print(f"Missing pada Sub-Category: {missing_sub} baris")

if missing_order == 0 and missing_sub == 0:
    print("✅ Tidak ada data hilang pada kolom transaksi & item")

# ============================================
# 4. CEK DATA DUPLIKAT
# ============================================
print("\n【4. CEK DATA DUPLIKAT】")
print("-"*40)

# Cek duplikat per baris
duplikat_baris = df.duplicated().sum()
print(f"Duplikat per baris: {duplikat_baris} baris")

# Cek duplikat transaksi-item (Order ID + Sub-Category)
duplikat_transaksi_item = df[['Order ID', 'Sub-Category']].duplicated().sum()
print(f"Duplikat (Order ID + Sub-Category): {duplikat_transaksi_item} pasangan")

if duplikat_transaksi_item == 0:
    print("✅ Tidak ada transaksi dengan item duplikat")
else:
    print(f"⚠️ Ada {duplikat_transaksi_item} pasangan duplikat yang perlu dibersihkan")

# ============================================
# 5. CEK VARIASI ITEM
# ============================================
print("\n【5. CEK VARIASI ITEM (Sub-Category)】")
print("-"*40)

jumlah_sub = df['Sub-Category'].nunique()
print(f"Jumlah sub-kategori unik: {jumlah_sub}")

# Daftar sub-kategori
sub_list = df['Sub-Category'].unique()
print(f"Daftar sub-kategori: {sorted(sub_list)}")

if jumlah_sub >= 2:
    print(f"✅ Terdapat {jumlah_sub} sub-kategori (>= 2, cukup untuk analisis)")
else:
    print("⚠️ Jumlah sub-kategori kurang dari 2, tidak bisa analisis asosiasi")

# ============================================
# 6. CEK VARIASI TRANSAKSI
# ============================================
print("\n【6. CEK VARIASI TRANSAKSI (Order ID)】")
print("-"*40)

jumlah_transaksi = df['Order ID'].nunique()
print(f"Jumlah transaksi unik (Order ID): {jumlah_transaksi}")

if jumlah_transaksi >= 100:
    print(f"✅ Transaksi cukup ({jumlah_transaksi} transaksi)")
else:
    print(f"⚠️ Jumlah transaksi sedikit ({jumlah_transaksi}), hasil mungkin kurang signifikan")

# ============================================
# 7. CEK ITEM PER TRANSAKSI
# ============================================
print("\n【7. CEK ITEM PER TRANSAKSI】")
print("-"*40)

item_per_transaksi = df.groupby('Order ID')['Sub-Category'].nunique()
stat_item = item_per_transaksi.describe()

print("Statistik jumlah item per transaksi:")
print(stat_item)

# Cek transaksi dengan 1 item
transaksi_1_item = (item_per_transaksi == 1).sum()
persen_1_item = (transaksi_1_item / len(item_per_transaksi)) * 100
print(f"\nTransaksi dengan 1 item: {transaksi_1_item} ({persen_1_item:.1f}%)")

# Cek transaksi dengan minimal 2 item
transaksi_2_item = (item_per_transaksi >= 2).sum()
persen_2_item = (transaksi_2_item / len(item_per_transaksi)) * 100
print(f"Transaksi dengan >= 2 item: {transaksi_2_item} ({persen_2_item:.1f}%)")

if transaksi_2_item >= 10:
    print("✅ Ada cukup transaksi dengan minimal 2 item untuk analisis asosiasi")
else:
    print("⚠️ Transaksi dengan >= 2 item sedikit, analisis mungkin tidak optimal")

# ============================================
# 8. CEK FREKUENSI ITEM
# ============================================
print("\n【8. CEK FREKUENSI ITEM】")
print("-"*40)

freq_sub = df['Sub-Category'].value_counts()
print("Top 10 sub-kategori paling sering muncul:")
print(freq_sub.head(10))

min_freq = freq_sub.min()
print(f"\nFrekuensi terendah: {min_freq} transaksi")

if min_freq >= 5:
    print(f"✅ Semua sub-kategori muncul minimal {min_freq} kali")
else:
    print(f"⚠️ Ada sub-kategori dengan frekuensi < 5, mungkin tidak bisa muncul di frequent itemsets")

# ============================================
# 9. CEK JUMLAH TRANSAKSI PER ITEM
# ============================================
print("\n【9. CEK SUPPORT AWAL】")
print("-"*40)

support_awal = freq_sub / jumlah_transaksi
print("Support 5 sub-kategori teratas:")
print(support_awal.head(5).round(4))

min_support_efektif = 5 / jumlah_transaksi
print(f"\nSupport minimal 5 transaksi: {min_support_efektif:.4f} ({min_support_efektif*100:.2f}%)")
print("Nilai ini bisa digunakan sebagai minimum support")

# ============================================
# 10. CEK MATRIKS BINARY (Contoh)
# ============================================
print("\n【10. CEK MATRIKS BINARY (Contoh 5 transaksi pertama)】")
print("-"*40)

# Buat matriks binary contoh
df_basket = df[['Order ID', 'Sub-Category']].drop_duplicates()
basket_sample = pd.crosstab(
    df_basket['Order ID'].head(5), 
    df_basket['Sub-Category']
)
print("Contoh matriks binary (5 transaksi pertama):")
print(basket_sample)

# ============================================
# 11. REKOMENDASI & KESIMPULAN
# ============================================
print("\n" + "="*60)
print("【REKOMENDASI & KESIMPULAN】")
print("="*60)

masalah = []
if not ada_semua:
    masalah.append("❌ Tidak ada kolom Order ID dan Sub-Category")
if missing_order > 0 or missing_sub > 0:
    masalah.append(f"⚠️ Ada {missing_order + missing_sub} data hilang")
if duplikat_transaksi_item > 0:
    masalah.append(f"⚠️ Ada {duplikat_transaksi_item} duplikat transaksi-item")
if jumlah_sub < 2:
    masalah.append("❌ Jumlah sub-kategori kurang dari 2")
if transaksi_2_item < 10:
    masalah.append(f"⚠️ Transaksi >=2 item hanya {transaksi_2_item} dari {len(item_per_transaksi)}")

if not masalah:
    print("\n✅✅✅ DATASET SIAP DIGUNAKAN UNTUK ANALISIS ASOSIASI ✅✅✅")
    print("\nRekomendasi parameter awal:")
    print(f"  - Min support: {min_support_efektif:.4f} (~5 transaksi)")
    print("  - Min confidence: 0.10 (untuk awal), 0.80 (untuk aturan kuat)")
    print("  - Algoritma: Apriori atau FP-Growth")
else:
    print("\n⚠️ Masalah yang ditemukan:")
    for m in masalah:
        print(f"  {m}")
    print("\nLangkah perbaikan yang perlu dilakukan:")
    if 'Order ID' not in df.columns or 'Sub-Category' not in df.columns:
        print("  1. Pastikan dataset memiliki kolom transaksi dan item")
    if missing_order > 0 or missing_sub > 0:
        print("  2. Hapus atau imputasi data hilang")
    if duplikat_transaksi_item > 0:
        print("  3. Hapus duplikat dengan drop_duplicates()")
    if transaksi_2_item < 10:
        print("  4. Pertimbangkan agregasi ke level item lain (Category, Product Name)")

# ============================================
# 12. VISUALISASI DISTRIBUSI ITEM PER TRANSAKSI
# ============================================
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
item_per_transaksi.hist(bins=range(1, item_per_transaksi.max() + 2), 
                        edgecolor='black', alpha=0.7)
plt.title('Distribusi Jumlah Sub-Category per Transaksi')
plt.xlabel('Jumlah Sub-Category')
plt.ylabel('Jumlah Transaksi')
plt.axvline(item_per_transaksi.median(), color='red', linestyle='--', 
            label=f'Median = {item_per_transaksi.median():.0f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
